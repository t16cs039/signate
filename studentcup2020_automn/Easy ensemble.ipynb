```
import pandas as pd

df_nn = pd.read_csv(PATH_NN, header=None)
df_lstm = pd.read_csv(PATH_LSTM, header=None)
df_bert = pd.read_csv(PATH_BERT, header=None)

df_nn.columns = ['id', 'sigmoid']
df_lstm.columns = ['id', 'sigmoid']
df_bert.columns = ['id', 'sigmoid']
```

# BERT + NN

```
df_nn['sigmoid'] = df_nn['sigmoid']*0.5
df_bert['sigmoid'] = df_bert['sigmoid']*0.5
df = pd.DataFrame(df_nn['id'], columns=['id'])
df['result'] = df_nn['sigmoid']+df_bert['sigmoid']
df['result'] = [0 if i < 0.5  else 1 for i in df['result'].tolist()]

df.to_csv('submission.csv', header=False, index=False)

df.head()
```

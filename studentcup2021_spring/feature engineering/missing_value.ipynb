```python
# for pandas
def standardization(data, target): 
  
  df_temp = (data - target.mean()) / target.std()

  return df_temp
```

# Preprocessing

## tempo

```python
df_temp = df['tempo'].str.split('-', expand=True)
df_temp.columns = ['tempo_min', 'tempo_max']
df_temp = df_temp.astype(int)

df = pd.concat([df, df_temp], axis=1)

df.head()
```

## one-hot-*encoding*

```python
ohe_list = ['popularity', 'tempo', 'region']
for col in ohe_list:
  df = pd.concat([df, one_hot_encoding(df[col], df[col])], axis=1)
  
print('df_train shape: {0}'.format(df.shape))
```

## 正規化

```python
ss_list = ['popularity', 'duration_ms', 'loudness', 'tempo_min', 'tempo_max']

for col in ss_list:
  df[col] = standardization(df[col], df[col])

print('df_train shape: {0}'.format(df.shape))
```

## Deleting wasted columns 

```python
df = df.drop(['index', 'tempo', 'region'], axis=1)
```
